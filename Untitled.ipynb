```python
df = df.sort_values(['marksObtained', 'advanceSection', 'Mathematics', 'Science', 'logicalReasoning', 'negativeAttempts', 'positiveAttempts'], ascending=[False, False, False, False, False, False, True])
df.index = np.arange(4, len(df)+4)
df['nationalRank'] = df.index
df['percentile'] = (((len(df)+4)-df.nationalRank)/(len(df)+4))*100
    
df = df.sort_values(['logicalReasoning'], ascending=[ False])
df.index = np.arange(4, len(df)+4)
df['logicalRank'] = df.index
df['logicalPercentile'] = (((len(df)+4)-df.logicalRank)/(len(df)+4))*100

df = df.sort_values(['Science'], ascending=[ False])
df.index = np.arange(4, len(df)+4)
df['scienceRank'] = df.index
df['sciencePercentile'] = (((len(df)+4)-df.scienceRank)/(len(df)+4))*100

df = df.sort_values(['Mathematics'], ascending=[ False])
df.index = np.arange(4, len(df)+4)
df['mathematicsRank'] = df.index
df['mathematicsPercentile'] = (((len(df)+4)-df.mathematicsRank)/(len(df)+4))*100

df = df.sort_values(['advanceSection'], ascending=[ False])
df.index = np.arange(4, len(df)+4)
df['advanceRank'] = df.index
df['advancePercentile'] = (((len(df)+4)-df.advanceRank)/(len(df)+4))*100
```

```python
df['nationalHighestLogical'] = df.logicalReasoning.max()
df['nationalHighestScience'] = df.Science.max()
df['nationalHighestMathematics'] = df.Mathematics.max()
df['nationalHighestAdvance'] = df.advanceSection.max()

df["nationalAverage"] = sum(df["marksObtained"])/len(df)

df = df.sort_values(['nationalRank'], ascending=[True])
df.index = np.arange(len(df))
```

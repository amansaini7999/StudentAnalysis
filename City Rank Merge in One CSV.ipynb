```python
import pandas as pd
import os
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
path = "/home/aman/Documents/eurekaGurus/Dataset/City Rank/01/"
```

```python
files = []
for r,d,f in os.walk(path):
    for file in f:
        files.append(os.path.join(r, file))
```

```python
df = pd.DataFrame(columns = ['Roll Number', 'City Rank'])
```

```python
for i in files:
    df1 = pd.read_csv(i)
    df = pd.concat([df, df1])
```

```python
df = df.rename(columns = {'Roll Number' : 'Key'})
df = df.set_index('Key')
df = df.sort_index()
df
```

```python
df.to_csv('/home/aman/Documents/eurekaGurus/Dataset/City Rank/Class 1.csv')
```

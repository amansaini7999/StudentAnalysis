```python
import pandas as pd
import os
import numpy as np
```

```python
pathKey = "/home/aman/Documents/eurekaGurus/Dataset/Answer Key/"
pathStudent = "/home/aman/Documents/eurekaGurus/Dataset/Solution/"
```

```python
files = []
for r,d,f in os.walk(pathKey):
    for file in f:
        files.append(os.path.join(r, file))
        
files.sort()

df1 = pd.read_csv( pathKey + files[0][52:], header = None)
```

```python
files = []
for r,d,f in os.walk(pathStudent):
    for file in f:
        files.append(os.path.join(r, file))
```

```python
c = 0
w = []

for i in range(12000):
    if files[i][57:59] == '01':
        w.append(files[i])
        c = c+1
```

```python
li = []

def Score():
    for i in range(c):
        
        df2 = pd.read_csv(pathStudent + w[i][50:], header = None)
        str = w[i][50:-4]
        str = str.split('/')
        s = "-"
        str = s.join(str)
        
        LR = 0
        SC = 0
        MH = 0
        AS = 0
        
        for j in range(0,10):
            if (df2[0][j] == -1):
                LR = LR+0
            elif (df1[0][j] == df2[0][j]):
                LR = LR+1
            else:
                LR = LR-0.25
        
        for j in range(10,20):
            if (df2[0][j] == -1):
                SC = SC+0
            elif (df1[0][j] == df2[0][j]):
                SC = SC+1
            else:
                SC = SC-0.25
                
        for j in range(20, 30):
            if (df2[0][j] == -1):
                MH = MH+0
            elif (df1[0][j] == df2[0][j]):
                MH = MH+1
            else:
                MH = MH-0.25
                
        for j in range(30, 35):
            if (df2[0][j] == -1):
                AS = AS+0
            elif (df1[0][j] == df2[0][j]):
                AS = AS+2
            else:
                AS = AS-0.50
                
        Marks = LR + SC + MH + AS

        li.append([str, LR, SC, MH, AS, Marks])
```

```python
Score()
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df = pd.DataFrame(li)
```

```python
df = df.rename(columns = {0:'Roll Number', 1:'Logical Reasoning', 2:'Science', 3:'Mathematics', 4:'Advance Section', 5:'Marks Obtained'})
df
```

```python
df['Percentage'] = (df['Marks Obtained']/40)*100
df['Percentage'] = df['Percentage'].round(2)
df['Percentile'] = (df['Marks Obtained']/df['Marks Obtained'].max())*100
df = df.sort_values(by='Percentage', ascending=0)
df['Percentile'] = df['Percentile'].round(2)
df
```

```python
df.index = np.arange(1, len(df)+1)
```

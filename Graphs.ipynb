```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("Class 1.csv")
df = df.set_index(['Key'])
df.head()
```

```python
df.columns
```

```python
LR = df['Logical Reasoning']
LR[0]

SC = df['Science']
SC[0]

Maths = df['Mathematics']
Maths[0]

AS = df['Advance Section']
AS[0]

M = df['Marks Obtained']
M[0]

NAM = df['Average Marks Obtained - Marks Obtained']
NAM[0]

CAM = df['Average Marks Obtained in City - Marks Obtained']
CAM[0]

SAM = df['Average Marks Obtained in School - Marks Obtained']
SAM[0]

attemptLR = df["Attempted Question in Logical Reasoning"]
attemptMaths = df["Attempted Question in Maths"]
attemptScience = df["Attempted Question in Science"]
attemptAS = df["Attempted Question in Advance Section"]

totalAttempt = df["Total Attempted Questions"]
notAttempt = 35-totalAttempt
negativeAttempted = df["Total Negative Attempted Questions"]
positiveAttempt = totalAttempt - negativeAttempted

totalLR = 10
totalMaths = 10
totalScience = 10
totalAS = 5

totalQuestion = 35
```

```python
plt.axis("equal")
plt.pie([LR[0], SC[0],Maths[0], AS[0]], labels = ["Logical Reasonig", "Science", "Maths", "Advance Section"], radius=1, autopct='%0.2f%%', explode=[0.1, 0.1, 0.1, 0.1])
plt.show()
```

```python
plt.bar(['Logical Reasoning','Science','Mathematics','Advance Section'],[LR[0],SC[0],Maths[0],AS[0]])
plt.show()
plt.plot(['Your Marks','School','City','National'],[M[0],SAM[0],CAM[0],NAM[0]], marker='*')
plt.show()
plt.bar(['Your Marks','School','City','National'],[M[0],SAM[0],CAM[0],NAM[0]])
plt.show()
plt.axis("equal")
plt.pie([LR[0], SC[0],Maths[0], AS[0]], labels = ["Logical Reasonig", "Science", "Maths", "Advance Section"])
plt.show()
plt.pie([attemptLR[0], attemptMaths[0], attemptScience[0], attemptAS[0]], labels = ["Attempted Questions in Logical Reasoning", "Attempted Questions in Mathematics", "Attempted Questions in Science", "Attempted Questions in Advance Section"])
plt.show()
plt.pie([totalAttempt[0], notAttempt[0]], labels = ["Total Attempted Questions", "Total Not Attempted Questions"])
plt.show()
plt.pie([positiveAttempt[0], negativeAttempted[0]], labels = ["Total Positive Attempted Questions", "Total Naegative Attempted Questions"])
plt.show()
```

```python
def result_view(key):
    print(key)
    plt.bar(['Logical Reasoning','Science','Mathematics','Advance Section'],[LR[key],SC[key],Maths[key],AS[key]])
    plt.show()
    plt.plot(['Your Marks','School','City','National'],[M[key],SAM[key],CAM[key],NAM[key]], marker='*')
    plt.show()
    plt.bar(['Your Marks','School','City','National'],[M[key],SAM[key],CAM[key],NAM[key]])
    plt.show()
    plt.axis("equal")
    plt.pie([LR[key], SC[key],Maths[key], AS[key]], labels = ["Logical Reasonig", "Science", "Maths", "Advance Section"])
    plt.show()
    plt.pie([attemptLR[key], attemptMaths[key], attemptScience[key], attemptAS[key]], labels = ["Attempted Questions in Logical Reasoning", "Attempted Questions in Mathematics", "Attempted Questions in Science", "Attempted Questions in Advance Section"])
    plt.show()
    plt.pie([totalAttempt[key], notAttempt[key]], labels = ["Total Attempted Questions", "Total Not Attempted Questions"])
    plt.show()
    plt.pie([positiveAttempt[key], negativeAttempted[key]], labels = ["Total Positive Attempted Questions", "Total Naegative Attempted Questions"])
    plt.show()
```

```python
result_view(0)
```

```python
result_view('C12001-01-0001')
```

```python
import pandas as pd
import os
import numpy as np
```

```python
pathKey = "/home/aman/Documents/eurekaGurus/Dataset/Answer Key/"
pathStudent = "/home/aman/Documents/eurekaGurus/Dataset/City 1/C12001/01/"
```

```python
files = []
for r,d,f in os.walk(pathKey):
    for file in f:
        files.append(os.path.join(r, file))
        
files.sort()

df1 = pd.read_csv( pathKey + files[0][52:], header = None)
```

```python
c = 0
files = []
for r,d,f in os.walk(pathStudent):
    for file in f:
        files.append(os.path.join(r, file))
        c = c+1
```

```python
li = []

def Score():
    for i in range(c):
        
        df2 = pd.read_csv(pathStudent + files[i][58:], header = None)
        str = files[i][48:-4]
        str = str.split('/')
        s = "-"
        str = s.join(str)
        
        Marks = 0
        
        for j in range(0,30):
            if (df2[0][j] == -1):
                Marks = Marks+0
            elif (df1[0][j] == df2[0][j]):
                Marks = Marks + 1
            else:
                Marks = Marks-0.25
                
        for j in range(30, 35):
            if (df2[0][j] == -1):
                Marks = Marks+0
            elif (df1[0][j] == df2[0][j]):
                Marks = Marks+2
            else:
                Marks = Marks-0.50

        li.append([str, Marks])
        
Score()
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df = pd.DataFrame(li)
```

```python
df = df.rename(columns = {0:'Roll Number', 1:'Marks Obtained'})
df
```

```python
df['Percentage'] = (df['Marks Obtained']/40)*100
df['Percentage'] = df['Percentage'].round(2)
df = df.sort_values(by='Percentage', ascending=0)
df
```

```python
df.index = np.arange(1, len(df)+1)
df
```

```python
df['Class Rank'] = df.index
df
```

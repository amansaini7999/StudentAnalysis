```python
df = pd.read_csv(path + "Ca")
df = df.fillna(-1)
df
```

```python
df.Subject[2]
```

```python
li = []

for i in range(len(df)):
    
    if df.Subject[i] == 'Maths':
        key = keyM
    else:
        key = keyB
    
    rollNumber = df.iloc[i][0]
    
    pCount = 0
    nCount = 0
    
    Marks = 0
    LR = 0
    SC = 0
    MH = 0
    AS = 0
    
    for j in range(10):
        if(key[j] == "BONUS"):
            LR = LR+1
        elif (df.iloc[i][j+1] == -1):
            LR = LR+0
        elif (df.iloc[i][j] == key[j]):
            LR = LR+1
            pCount = pCount+1
        else:
            LR = LR-0.25
            nCount = nCount+1
            
    for j in range(10, 20):
        if(key[j] == "BONUS"):
            SC = SC+1
        elif (df.iloc[i][j+1] == -1):
            SC = SC+0
        elif (df.iloc[i][j] == key[j]):
            SC = SC+1
            pCount = pCount+1
        else:
            SC = SC-0.25
            nCount = nCount+1
             
    for j in range(20, 30):
        if(key[j] == "BONUS"):
            MH = MH+1
        elif (df.iloc[i][j+1] == -1):
            MH = MH+0
        elif (df.iloc[i][j] == key[j]):
            MH = MH+1
            pCount = pCount+1
        else:
            MH = MH-0.25
            nCount = nCount+1
                   
    for j in range(30, 35):
        if(key[j] == "BONUS"):
            AS = AS+2
        elif (df.iloc[i][j+1] == -1):
            AS = AS+0
        elif (df.iloc[i][j] == key[j]):
            AS = AS + 2
            pCount = pCount+1
        else:
            AS = AS-0.50
            nCount = nCount+1
            
    Marks = LR + SC + MH + AS         
    li.append([rollNumber,LR, SC, MH, AS, Marks, pCount, nCount])
```

```python
df = pd.DataFrame(li)
df
```

```python
df = df.rename(columns = {0:'rollNumber', 1:'logicalReasoning', 2:'Science', 3:'Mathematics', 4:'advanceSection', 5:'marksObtained', 6:'positiveAttempts', 7:'negativeAttempts'})
df
```
